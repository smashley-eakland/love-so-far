```python
import pandas as pd
import folium
```

```python
locations = pd.read_excel('./Data/Locations.xlsx')
```

```python
locations.drop(columns=['City.1'], inplace=True)
```

```python
locations
```

```python
locations.dropna(subset=['City'], inplace=True)
locations
```

```python
#Kerry Park starting location
base_map = folium.Map([47.629486, -122.359909], zoom_start=1)
base_map
```

```python
#initializing marker variables
long = locations['Longitude']
lat = locations['Latitude']
name = locations['City']
points = list(zip(lat, long, name))
```

```python
#adding markers with popups to base map
for p in points:
    lat = p[0]
    long = p[1]
    name = p[2]
    popup = folium.Popup(name)
    marker = folium.Marker(location=[lat,long], popup=popup)
    marker.add_to(base_map)
```

```python
base_map
```

```python
from geopy.distance import geodesic
from geopy import distance
```

```python
#calculate distance from Seattle
seattle = (47.629486, -122.359909)
long = locations['Longitude']
lat = locations['Latitude']
coords = list(zip(lat, long))
distances = []
for loc in coords:
    distances.append(distance.distance(loc, seattle))
```

```python
#add it to the DF
locations['Distance'] = distances
```

```python
locations
```

```python
#reformat our base map to include location distance from Seattle "home base"
distances = locations['Distance']
long = locations['Longitude']
lat = locations['Latitude']
name = locations['City']
points = list(zip(lat, long, name, distances))
for p in points:
    lat = p[0]
    long = p[1]
    name = p[2]
    dist_frm = p[3]
    dist_pop = f' {name} : {dist_frm} from Seattle!'
    popup = folium.Popup(dist_pop)
    marker = folium.Marker(location=[lat,long], popup=popup)
    marker.add_to(base_map)
    
base_map
```

```python
locations['Distance from Seattle'] = locations['Distance'].copy().astype(str)
```

```python
locations['Distance from Seattle'] = locations['Distance from Seattle'].str.replace(" km", "").astype(float)
```

```python
#change some stylings!
distances = locations['Distance from Seattle']
long = locations['Longitude']
lat = locations['Latitude']
name = locations['City']
points = list(zip(lat, long, name, distances))
for p in points:
    lat = p[0]
    long = p[1]
    name = p[2]
    dist_frm = p[3]
    dist_pop = f' {name} : {dist_frm:.2f}km from Seattle!'
    popup = folium.Popup(dist_pop)
    icon = folium.Icon(color='red', icon_color='white', icon='heart', angle=0, prefix='fa')
    marker = folium.Marker(location=[lat,long], popup=popup, icon=icon)
    marker.add_to(base_map)
    
base_map
```
